  已知一个二次函数 y = x² - 2x - 3，请：
   1. 绘制这个函数的图像。
   2. 找出它的顶点坐标。
   3. 找出它与 x 轴的交点（即方程的根）。

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 定义我们的二次函数
def parabola(x):
   return x**2 - 2*x - 3

# 生成 x 值，例如从 -5 到 7，共 100 个点
x = np.linspace(-5, 7, 100)
    
# 计算对应的 y 值
y = parabola(x)
```

```python
# 创建图像
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x² - 2x - 3')

# 添加辅助线和标签
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.title('Plot of the Parabola y = x² - 2x - 3')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()

# 显示图像
plt.show()
```
